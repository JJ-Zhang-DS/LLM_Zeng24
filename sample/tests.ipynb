```python
def call_openai_chat(model, messages):
    result = client.chat.completions.create(
        messages=messages,
        model=model,
    )
    return result.choices[0].message.content

messages = [
    {
        "role": "user",
        "content": "Say this is a test",
    },
    {
        "role": "system",
        "content": "You are a helpful assistant.",
    },
    {
        "role": "user",
        "content": "who is the us president in 2004?"
    },
    {
        "role": "system",
        "content": "George W. Bush was the president of the United States in 2004."
    },
    {
        "role": "user",
        "content": "who comes after him"
    }

]
```

```python
call_openai_chat("gpt-4o", messages)
```
